# Data on a Company GG (not disclosed) that Gives Delayed Salary

There is a certain company that we'll call as GG.

- They have been in business and lost employees in the past.

NOTE: Data here is generated by random due to the fact that it is very sensitive for the employees to ask these information from their immature management team.

- this is not fictional, however the exact number of questions asked cannot be asked of. For the sake of learning visualization with animation this suits our needs.

```python
# Importing required libraries

import bar_chart_race as bcr
import pandas as pd
```

```python
df = pd.read_csv('data/data_delay_viz.csv')
```

```python
print(df.info()) 
df.head(3)
```

```python
from datetime import date
from datetime import datetime

df['date'] = df['date'].apply(
    lambda x: datetime.strptime(x, "%m/%d/%Y").date())
```

```python
df['date'].head()
```

```python
df.head()
```

```python
df.columns = ["date", "applicants", "resigning", "asking_delayed_salary"]
```

```python
df.head()
```

```python
df['applicants'] = df['applicants'].cumsum()
```

```python
df['resigning'] = df['resigning'].cumsum()
```

```python
df['asking_delayed_salary'] = df['asking_delayed_salary'].cumsum()
```

```python
df.head()
```

```python
df.set_index("date", inplace = True)
```

```python
df.head()
```

```python
bcr.bar_chart_race(df=df,
                  n_bars=3,
                  sort='desc',
                  fixed_max = True,
                  bar_size=.7,
                  bar_kwargs={'alpha': .7},
                  fig_kwargs={'figsize': (6, 3.5), 'dpi': 144},
                  period_label={'x': .98, 'y': .3, 'ha': 'right', 'va': 'top'},
                  bar_textposition='outside',
                  bar_texttemplate='{x:,.0f}',
                  bar_label_font=7, 
                  tick_label_font=7, 
                  tick_template='{x:,.0f}',
                  img_label_folder='bar_image_labels_comp',
                  title='Cummulative Count of Employees in GG (fictional name)',
                  period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': 'by: @wenmi01 \n from: random data GG',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'})
```
